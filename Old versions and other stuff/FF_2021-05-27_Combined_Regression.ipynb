# Applying a Regresion to the overall scores of the lexical metrics and embeddings

```python
import os
import pandas as pd
import numpy as np
from scipy.stats import zscore
from sklearn.linear_model import LinearRegression

lexical_path = "/Users/franz/Downloads/Text_Mining_Group_Project-main/Franz (lexical) - csvs with predictions/train/"
embeddings_path = "/Users/franz/Downloads/Text_Mining_Group_Project-main/Philipp (WM's distance) - csvs with predictions/train/"
```

```python
lexical_train = os.listdir(lexical_path)
embeddings_train = os.listdir(embeddings_path)
```

```python
print(lexical_train)
```

```python
print(embeddings_train)
```

## RU - EN

```python
kendalls = []
pearsons = []
p_e = []
p_l = []
k_e = []
k_l = []
```

```python
l = pd.read_csv(lexical_path + 'ru_en_predictions.csv')
e = pd.read_csv(embeddings_path + 'train_all_with_predictions_ru_en.csv')
```

```python
l = l[["prediction"]].rename(columns={"prediction":"prediction_l"})
e = e[["z-score", "prediction"]].rename(columns={"prediction":"prediction_e"})
ru_en = e
ru_en["prediction_l"] = l["prediction_l"]
```

```python
p = ru_en.corr()
display(p)
p_e.append(p.iloc[1,0])
p_l.append(p.iloc[2,0])

k = ru_en.corr(method="kendall")
display(k)
k_e.append(k.iloc[1,0])
k_l.append(k.iloc[2,0])
```

```python
for element in ["prediction_e", "prediction_l"]:
    ru_en[element] = zscore(ru_en[element])
    
ru_en__train = ru_en.iloc[:int(ru_en.shape[0]*0.7),:]
ru_en__test = ru_en.iloc[int(ru_en.shape[0]*0.7):,:]

X = ru_en__train.iloc[:,1:]
Y = ru_en__train.iloc[:,0]
X_test = ru_en__test.iloc[:,1:]
Y_test = ru_en__test.iloc[:,0] 
X_full = ru_en.iloc[:,1:]
Y_full = ru_en.iloc[:,0] 

model_ru_en = LinearRegression()
model_ru_en.fit(X,Y)

final_model_ru_en = LinearRegression()
final_model_ru_en.fit(X_full, Y_full)
    
result = pd.DataFrame(Y_test)
result["regression_values"] = model_ru_en.predict(X_test)

p = result.corr()
display(p)
pearsons.append(p.iloc[1,0])

k = result.corr(method="kendall")
display(k)
kendalls.append(k.iloc[1,0])
```
